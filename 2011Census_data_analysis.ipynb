```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Census+2011.csv')
df.columns  #displays all columns
```

```python
df.head(2) #displays the top n rows, similarly you can use tail for the last n rows
```

```python
df.style.set_caption("Census_2011") #style the table with the new namings
df.head(3)
```

```python
df[df['District_name'].isin(['Lucknow','Jaipur'])] #isin operator is used to search for the corresponding value in a particular column
```

```python
df.groupby('State_name').Population.sum() #when u want multiple values of as single column you can use groupby
```

```python
df.groupby('State_name').Population.sum().sort_values()
```

```python
df.groupby('State_name')[[ 'Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum() #*** use double[[]] when u always want the ans for the multiple values after groupby
```

```python
df.groupby('State_name')[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum().sort_values(by='Hindus' ,ascending = False).head(5)
```

```python
df[df.State_name == 'MAHARASHTRA']['Male_Workers'].sum() #when u need the ans for a single value of a particular column no need to use groupby
```

```python
kar_men_workers = df[df.State_name == 'KARNATAKA']['Male_Workers'].sum()
print(kar_men_workers)
```

```python
kar_female_workers = df[df.State_name == 'KARNATAKA']['Female_Workers'].sum()
print(kar_female_workers)
```

```python
sizes = [kar_men_workers, kar_female_workers]  # combined data for men and women
labels = ['Men', 'Women']  # labels for the slices
colors = ['blue', 'pink']  # colors for the slices

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')

# Add title
plt.title('Male vs Female Employment')

plt.legend()
# Show the plot
plt.show()
```
